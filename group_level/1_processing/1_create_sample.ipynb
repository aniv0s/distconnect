```python
age_max = 40
print 'N =', df_lsd.subject[(df_lsd['age'] <= age_max)].shape[0], 'using maximum age of', age_max
```

### quality metrics

```python
# checking overall quality of scans in remaining sample

for metric in ['maxFD', 'meanFD', 'coreg', 'tSNR']:
    g = sns.FacetGrid(df_tidy[df_tidy.subject.isin(df_lsd.subject.tolist())], col='scan')
    g.map(plt.hist, metric)
```

### checking relationship between quality metrics and nan occurance, before excluding based on motion cutoff 

```python
for metric in ['maxFD', 'meanFD', 'coreg']:
    for hemi in ['lh', 'rh']:
        g = sns.FacetGrid(df_tidy[df_tidy.subject.isin(df_lsd.subject.tolist()) & (df_tidy['surf_raw_%s_nan_count' % hemi] > 0)], col='scan')
        g.map(plt.scatter, metric, 'surf_raw_%s_nan_count' % hemi, alpha=.7)
```

### effect of motion on coregistration, , before excluding based on motion cutoff 

```python
# exploring relationship between head motion and preprocessing quality
for metric in ['maxFD', 'meanFD']:
    for hemi in ['lh', 'rh']:
        g = sns.FacetGrid(df_tidy[df_tidy.subject.isin(df_lsd.subject.tolist())], col='scan')
        g.map(plt.scatter, metric, 'coreg', alpha=.7)
```
