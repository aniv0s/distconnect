# sample for analysis

```python
import pandas as pd
import os, glob,  time, datetime
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='ticks')

# paths, files
docDir = '/afs/cbs.mpg.de/projects/mar005_lsd-lemon-surf/documents'
overview_f = max(glob.iglob('%s/data_overview_2015*.pkl' % docDir), key=os.path.getctime)
overview_tidy_f = max(glob.iglob('%s/data_overview_tidy_*.pkl' % docDir), key=os.path.getctime)

# output
date = datetime.datetime.now().strftime("%Y%m%d")
sample_f = '%s/distconnect_sample_%s.csv' % (docDir, date)
sample_nonan_f = '%s/distconnect_sample_nonan_%s.csv' % (docDir, date)

# dataframes
df = pd.read_pickle(overview_f)
df_tidy = pd.read_pickle(overview_tidy_f)
```

#### available meanDist maps for all four scans and both hemispheres

```python
print 'N =', df.subject[(df['meanDist_1ab2ab_75_lh'] == 1) & (df['meanDist_1ab2ab_75_rh'] == 1)].shape
```

```python
df = df[(df['meanDist_1a_75_lh'] == 1)
       & (df['meanDist_1a_75_rh'] == 1)
       & (df['meanDist_1b_75_lh'] == 1)
       & (df['meanDist_1b_75_rh'] == 1)
       & (df['meanDist_2a_75_lh'] == 1)
       & (df['meanDist_2a_75_rh'] == 1)
       & (df['meanDist_2b_75_lh'] == 1)
       & (df['meanDist_2b_75_rh'] == 1)]

print 'N =', df.subject.shape
```

#### age range before exclusion

```python
g = sns.FacetGrid(df_tidy[df_tidy.subject.isin(df.subject.tolist()) & (df_tidy['scan']=='1a')], col='Gender')
g.map(plt.hist, 'Age', bins=20, range=(15,80))
```

#### maximum age

```python
age_max = 40
print 'N =', df.subject[(df['Age'] <= age_max)].shape, 'using maximum age of', age_max
```

```python
# apply cutoff
df = df[(df['Age'] <= age_max)]
```

```python
g = sns.FacetGrid(df_tidy[df_tidy.subject.isin(df.subject.tolist()) & (df_tidy['scan']=='1a')], col='Gender')
g.map(plt.hist, 'Age', bins=20, range=(15,50))
```

#### quality metrics

```python
for metric in ['maxFD', 'meanFD', 'coreg', 'tSNR']:
    g = sns.FacetGrid(df_tidy[df_tidy.subject.isin(df.subject.tolist())], col='scan')
    g.map(plt.hist, metric)
```

```python
# different cutoffs
for maxFD_cutoff in [1, 1.1, 1.2, 1.3, 1.4, 1.5]:
    print 'N =', df.subject[(df['maxFD_1a'] <= maxFD_cutoff)
                           & (df['maxFD_1b'] <= maxFD_cutoff)
                           & (df['maxFD_2a'] <= maxFD_cutoff)
                           & (df['maxFD_2b'] <= maxFD_cutoff)].shape, 'using maxFD cutoff of', maxFD_cutoff
# set for later    
maxFD_cutoff = 1
```

#### before excluding based on motion cutoff, checking relationship between quality metrics and nan occurance

```python
for metric in ['maxFD', 'meanFD', 'coreg']:
    for hemi in ['lh', 'rh']:
        g = sns.FacetGrid(df_tidy[df_tidy.subject.isin(df.subject.tolist()) & (df_tidy['surf_raw_%s_nan_count' % hemi] > 0)], col='scan')
        g.map(plt.scatter, metric, 'surf_raw_%s_nan_count' % hemi, alpha=.7)
```

#### effect of motion on coregistration

```python
for metric in ['maxFD', 'meanFD']:
    for hemi in ['lh', 'rh']:
        g = sns.FacetGrid(df_tidy[df_tidy.subject.isin(df.subject.tolist())], col='scan')
        g.map(plt.scatter, metric, 'coreg', alpha=.7)
```

```python
# apply motion cutoff
df = df[(df['maxFD_1a'] <= maxFD_cutoff)
       & (df['maxFD_1b'] <= maxFD_cutoff)
       & (df['maxFD_2a'] <= maxFD_cutoff)
       & (df['maxFD_2b'] <= maxFD_cutoff)]
```
