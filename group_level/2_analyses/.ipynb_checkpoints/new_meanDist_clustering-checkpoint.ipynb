```python
%matplotlib inline

from plotting import *
import pandas as pd, numpy as np, nibabel as nib, seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
import glob, os, time, datetime
from IPython.core.display import Image, display

sns.set_context('talk')
sns.set_style('white')

docDir = '/afs/cbs.mpg.de/projects/mar005_lsd-lemon-surf/documents'
annotDir = '/scr/liberia1/data/Yeo_JNeurophysiol11_FreeSurfer'
fsDir = '/afs/cbs.mpg.de/projects/mar004_lsd-lemon-preproc/freesurfer'
```

```python
# data
df = pd.read_pickle(max(glob.iglob('%s/data_grouplevel/lsd_data_grouplevel_20*.pkl' % docDir), 
                        key=os.path.getctime))
df = df[(df['sample'] == 'LSD subjects')
       & (df['threshold'] == '95')]
```

```python
df_eval = pd.read_pickle(max(glob.iglob('%s/data_grouplevel/gmm_evaluations_20*.pkl' % docDir), 
                        key=os.path.getctime))

df_eval = df_eval[df_eval['threshold'] == '95']
```

```python
for num_comp in [3]: #[2,3,4,5,6,7,8,9]:
    col_name = 'gmm_comp%s' % num_comp
    for hemi in ['rh', 'lh']:
        print '\n%s, %s components' % (hemi, num_comp)

        # sort regions by their average meanDists
        
        d = {}
        for region in set(df[col_name][df[col_name] != 0]):
            d[region] = df['mean distance (geodesic) - group mean'][(df['hemisphere'] == hemi)
                                                                    & (df[col_name] == region)].mean()
        region_order = [d.keys()[d.values().index(x)] for x in np.sort(d.values())]


        # plots
        data = np.array(df['mean distance (geodesic) - group mean'][df['hemisphere'] == hemi])


        region_labels = tuple([df['node'][(df[col_name] == region) 
                                     & (df['hemisphere'] == hemi)]
                             for region in region_order])

        pal = sns.color_palette("inferno", len(region_order))

        create_fig(labels=region_labels, hemi=hemi, surf='inflated', cpal=pal, cbar=True)
        create_fig(data=data,labels=region_labels, hemi=hemi, surf='inflated',
                   sulc=False, cmap='inferno', cbar=True)

    # ToDo's: split by hemi, but keep color per violin
    sns.violinplot(x=col_name, 
                   y='mean distance (geodesic) - group mean', 
                   data=df[(df['networks of Yeo & Krienen 2011 (17 networks solution)'] != 'FreeSurfer_Defined_Medial_Wall')],
                   split=False, 
                   palette=pal, order=region_order)
```
