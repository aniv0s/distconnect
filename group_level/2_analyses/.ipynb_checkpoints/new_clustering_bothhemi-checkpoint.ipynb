```python
from plotting import *
import pandas as pd, numpy as np, nibabel as nib, seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
import glob, os, time, datetime
from IPython.core.display import Image, display

sns.set_context('talk')
sns.set_style('white')

docDir = '/afs/cbs.mpg.de/projects/mar005_lsd-lemon-surf/documents'
annotDir = '/scr/liberia1/data/Yeo_JNeurophysiol11_FreeSurfer'
fsDir = '/afs/cbs.mpg.de/projects/mar004_lsd-lemon-preproc/freesurfer'
```

```python
# data
df = pd.read_pickle(max(glob.iglob('%s/data_grouplevel/lsd_data_grouplevel_20*.pkl' % docDir), 
                        key=os.path.getctime))
df = df[(df['sample'] == 'LSD subjects')
       & (df['threshold'] == '98')]

# combine data from two hemispheres
temp = []
for hemi in ['rh', 'lh']:
    cort = np.sort(nib.freesurfer.io.read_label('%s/fsaverage5/label/%s.cortex.label' % (fsDir, hemi)))
    print len(cort)
    t = np.array(df['mean distance (geodesic) - group mean'][df['hemisphere'] == hemi])[cort].tolist()
    temp = temp + t

data = np.array(temp)


# distance in mm
gr_dist = np.load('%s/data_grouplevel/lsd_geoDist_groupmean_lh.npy' % docDir)
renorm_lh = gr_dist.max()
gr_dist = np.load('%s/data_grouplevel/lsd_geoDist_groupmean_rh.npy' % docDir)
renorm_rh = gr_dist.max()

renorm = (renorm_lh + renorm_lh) / 2
data = data * renorm
```

```python
from sklearn import mixture


def run_clustering(n_comp, n_bins, d):    
    
    gmm = mixture.GMM(n_components=n_comp, n_iter=1000)
    gmm.fit(d[:, np.newaxis])
    res = gmm.predict(d[:, np.newaxis])
    res = res +1
    
    # save label files for each hemi
    
    comp_labels_rh = res[0:9361]
    comp_labels_lh = res[9361::]
    comp_labels_both = res
    
    labels_rh = np.zeros((10242))
    cort = np.sort(nib.freesurfer.io.read_label('%s/fsaverage5/label/%s.cortex.label' % (fsDir, 'rh')))
    labels_rh[cort] = comp_labels_rh
    np.save('/afs/cbs.mpg.de/projects/mar005_lsd-lemon-surf/results/clustering/clust_%icomp_rh' % n_comp, labels_rh)
    
    labels_lh = np.zeros((10242))
    cort = np.sort(nib.freesurfer.io.read_label('%s/fsaverage5/label/%s.cortex.label' % (fsDir, 'lh')))
    labels_lh[cort] = comp_labels_lh
    np.save('/afs/cbs.mpg.de/projects/mar005_lsd-lemon-surf/results/clustering/clust_%icomp_lh' % n_comp, labels_lh)
    
    # cutoff values
    #print data[np.where(comp_labels_both == n_comp)[0]].min()
    comp_mean = []
    for comp in range(1,n_comp+1):
        comp_mean.append(data[np.where(comp_labels_both == int(comp))[0]].mean())
    
    comp_mean = np.array(comp_mean)
    print comp_mean
    
    
    cutoffs = []
    for comp in range(1,n_comp+1):
        cutoffs.append(data[np.where(comp_labels_both == int(comp))[0]].min())
    print cutoffs
    print data.max()
    
    
    # plot grid
    
    bins = plt.hist(data, bins = n_bins)
    plt.close()
    
    sns.set_style("whitegrid")
    
    def gaussian(x, mu, sig):
        return np.exp(-np.power(x - mu, 2.) / (2 * np.power(sig, 2.)))
    
    order = np.argsort(comp_mean)+1
    #print comp_mean[order-1]
    
    for comp in order[::-1]:
        m = data[np.where(comp_labels_both == comp)[0]].mean()
        std = data[np.where(comp_labels_both == comp)[0]].std()
        
        bin_index = np.where(bins[1] == min(bins[1], key=lambda x:abs(m-x)))[0][0]
        scale = bins[0][bin_index]
        #scale = (bins[0][bin_index] + bins[0][bin_index-1] + bins[0][bin_index+1] + bins[0][bin_index-2] + bins[0][bin_index+2]
        #         + bins[0][bin_index-3] + bins[0][bin_index+3]) / 7
        
        values = gaussian(np.linspace(5, 110, 120), m, std) #+ scale #* maxi
        
        plt.plot(np.linspace(5, 110, 120), values * scale)
        
    sns.distplot(d, hist=True, kde=False, bins=n_bins)
    plt.axis((0,125, 0, 600))
    plt.xlabel('distance in mm')
    plt.ylabel('count')
```

```python
def plot_modelfit(data, n_comp, gmm):
    hist(data, bins=50, normed=1)
    x = np.linspace(min(data), max(data), 1000)
    for i in range(n_comp): 
        y = gmm.weights_[i]*stats.norm.pdf(x, loc=gmm.means_[i], scale=sqrt(gmm.covars_[i]))
        plot(x, y)
    #pyplot.savefig(('%s_hist.png' % filenamePrefix))
    plt.show()
    clf()
```
