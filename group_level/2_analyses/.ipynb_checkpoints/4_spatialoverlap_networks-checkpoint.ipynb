```python
%matplotlib inline

from plotting_lighterbg import *
import pandas as pd, numpy as np, nibabel as nib, seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
import glob, os, time, datetime
from IPython.core.display import Image, display

sns.set_context('talk')
sns.set_style('white')

docDir = '/afs/cbs.mpg.de/projects/mar005_lsd-lemon-surf/documents'
annotDir = '/scr/liberia1/data/Yeo_JNeurophysiol11_FreeSurfer'
fsDir = '/afs/cbs.mpg.de/projects/mar004_lsd-lemon-preproc/freesurfer'
```

```python
# data
df = pd.read_pickle(max(glob.iglob('%s/data_grouplevel/lsd_data_grouplevel_20*.pkl' % docDir), 
                        key=os.path.getctime))
df = df[df['sample'] == 'LSD subjects']


# group geo dist
gr_dist = np.load('%s/data_grouplevel/lsd_geoDist_groupmean_lh.npy' % docDir)
renorm_lh = gr_dist.max()

gr_dist = np.load('%s/data_grouplevel/lsd_geoDist_groupmean_rh.npy' % docDir)
renorm_rh = gr_dist.max()


# meanDist in mm

df_mm = df.copy()
renorm = (renorm_lh+renorm_rh)/2
df_mm['mean distance (geodesic) - group mean'] = df_mm['mean distance (geodesic) - group mean']* renorm



# yeo map
yeo7_names = ['medial wall', 'visual', 'somatomotor', 'dorsal attention', 
              'ventral attention', 'limbic', 'frontoparietal', 'default']
yeo_annot_f = '%s/fsaverage5/label/lh.Yeo2011_7Networks_N1000.annot' % (annotDir)
yeo7 = nib.freesurfer.io.read_annot(yeo_annot_f)
yeo7_pal = sns.color_palette(['#%02x%02x%02x' % tuple(color[:3]) for color in yeo7[1]])
```

## comparison to Yeo & Krienen FC networks (17 cluster solution)

```python
# yeo17 map
yeo17_names = ['medial wall','Vp','Vc','SMa','SMb','DAa','DAb','VA',
               'S','L','L','Cc','Ca','Cb','Dd','Dc','Da','Db']
yeo17_annot_f = '%s/fsaverage5/label/lh.Yeo2011_17Networks_N1000.annot' % (annotDir)
yeo17 = nib.freesurfer.io.read_annot(yeo17_annot_f)
yeo17_pal = sns.color_palette(['#%02x%02x%02x' % tuple(color[:3]) for color in yeo17[1]])

sns.palplot(yeo17_pal)
```

```python
for hemi in ['lh']:#['rh', 'lh']:
    print '\n%s' % hemi
    
    # sort networks by their average meanDists
    d = {}
    for netw in set(df['networks of Yeo & Krienen 2011 (17 networks solution)'][
                    (df['networks of Yeo & Krienen 2011 (17 networks solution)'] != 'FreeSurfer_Defined_Medial_Wall')]):
        d[netw] = df['mean distance (geodesic) - group mean'][(df['networks of Yeo & Krienen 2011 (17 networks solution)'] == netw)
                                                              & (df['threshold'] == '95')
                                                              & (df['hemisphere'] == hemi)].mean()
    
    netw_order = [d.keys()[d.values().index(x)] for x in np.sort(d.values())]
    pal_ordered = [yeo17_pal[yeo17[2].index(netw)] for netw in netw_order]

    
    # plots
    
    data = np.array(df['mean distance (geodesic) - group mean'][(df['threshold'] == '95') 
                                                     & (df['hemisphere'] == hemi)])
    
    
    netw_labels = tuple([df['node'][(df['networks of Yeo & Krienen 2011 (17 networks solution)'] == netw) 
                                 & (df['hemisphere'] == hemi)
                                 & (df['threshold'] == '95')]
                         for netw in netw_order])

    
    
    create_fig(labels=netw_labels, hemi=hemi, surf='inflated', cpal=pal_ordered, cbar=False, 
               file_name='/SCR/sabine/distconnect/figures/images/4_networks')
        
```

## network-specific distributions of connectivity distance

```python
thr = '98'

# sort networks by their average meanDists (hemispheres combined)
d = {}
for netw in set(df['networks of Yeo & Krienen 2011 (17 networks solution)'][
                (df['networks of Yeo & Krienen 2011 (17 networks solution)'] != 'FreeSurfer_Defined_Medial_Wall')]):
    d[netw] = df['mean distance (geodesic) - group mean'][(df['networks of Yeo & Krienen 2011 (17 networks solution)'] == netw)
                                                          & (df['threshold'] == thr)].median()

netw_order = [d.keys()[d.values().index(x)] for x in np.sort(d.values())]
pal_ordered = [yeo17_pal[yeo17[2].index(netw)] for netw in netw_order]
names_ordered = [yeo17_names[yeo17[2].index(netw)] for netw in netw_order]

# ToDo's: split by hemi, but keep color per violin
sns.set_context('poster')
sns.set(style="whitegrid")

f, ax = plt.subplots()
sns.violinplot(x='networks of Yeo & Krienen 2011 (17 networks solution)', 
               y='mean distance (geodesic) - group mean', 
               data=df_mm[(df_mm['networks of Yeo & Krienen 2011 (17 networks solution)'] != 'FreeSurfer_Defined_Medial_Wall')
                      & (df_mm['threshold'] == thr)],
               split=True, 
               palette=pal_ordered, order=netw_order,
              inner='quartile')

sns.despine(left=True, bottom=True)
ax.set_xticklabels(names_ordered)
ax.set_ylabel('distance in mm')
ax.set_xlabel('resting-state networks')
```

## create maps according to clustering

```python
df['network clusters'] = 0

df['network clusters'][df['networks of Yeo & Krienen 2011 (17 networks solution)']
                       .isin(netw_order[0:4])] = int(1)

df['network clusters'][df['networks of Yeo & Krienen 2011 (17 networks solution)']
                       .isin(netw_order[4:11])] = int(2)

df['network clusters'][df['networks of Yeo & Krienen 2011 (17 networks solution)']
                       .isin(netw_order[11:17])] = int(3)
```

```python
# iterating over thresholds and clustering based on kmeans


# clustering was obtained from k-means on pair-wise JS divergences of network distributions
clusterings = [[slice(0,3), slice(3,11), slice(11,17)],
               [slice(0,3), slice(3,11), slice(11,17)],
               [slice(0,4), slice(4,11), slice(11,17)],
               [slice(0,5), slice(5,11), slice(11,17)],
               [slice(0,5), slice(5,11), slice(11,17)],
               [slice(0,4), slice(4,11), slice(11,17)],
               [slice(0,4), slice(4,11), slice(11,17)]]


df['network clusters'] = 0
for n_thr, thr in enumerate(['70', '75', '80', '85', '90', '95', '98']):
    

    # sort networks by their average meanDists (hemispheres combined)
    d = {}
    for netw in set(df['networks of Yeo & Krienen 2011 (17 networks solution)'][
                    (df['networks of Yeo & Krienen 2011 (17 networks solution)'] != 'FreeSurfer_Defined_Medial_Wall')]):
        d[netw] = df['mean distance (geodesic) - group mean'][(df['networks of Yeo & Krienen 2011 (17 networks solution)'] == netw)
                                                              & (df['threshold'] == thr)].median()

    netw_order = [d.keys()[d.values().index(x)] for x in np.sort(d.values())]
    pal_ordered = [yeo17_pal[yeo17[2].index(netw)] for netw in netw_order]
    names_ordered = [yeo17_names[yeo17[2].index(netw)] for netw in netw_order]

    # visualize sorting with violins
    sns.set_context('poster')
    sns.set(style="whitegrid")

    f, ax = plt.subplots()
    sns.violinplot(x='networks of Yeo & Krienen 2011 (17 networks solution)', 
                   y='mean distance (geodesic) - group mean', 
                   data=df_mm[(df_mm['networks of Yeo & Krienen 2011 (17 networks solution)'] != 'FreeSurfer_Defined_Medial_Wall')
                          & (df_mm['threshold'] == thr)],
                   split=True, 
                   palette=pal_ordered, order=netw_order,
                  inner='quartile')

    sns.despine(left=True, bottom=True)
    ax.set_xlabel('Resting-state networks', fontsize=25)
    ax.set_xticklabels(names_ordered, fontsize=18)
    plt.xticks(rotation=45) 
    ax.set_ylabel('Connectivity distance (mm)', fontsize=25)
    plt.yticks(fontsize=18)
    
    f.savefig('/scr/hunte1/sabine/distconnect/figures/images/suppl_netw_violins_%s.png' 
              %thr, bbox_inches='tight', dpi=600)
    
    
    
    
    
    
    
    
    # assign nodes to cluster membership
    

    df['network clusters'][(df['threshold'] == thr) & 
                           (df['networks of Yeo & Krienen 2011 (17 networks solution)']
                            .isin(netw_order[clusterings[n_thr][0]]))] = int(1)

    df['network clusters'][(df['threshold'] == thr) & 
                           (df['networks of Yeo & Krienen 2011 (17 networks solution)']
                            .isin(netw_order[clusterings[n_thr][1]]))] = int(2)

    df['network clusters'][(df['threshold'] == thr) & 
                           (df['networks of Yeo & Krienen 2011 (17 networks solution)']
                            .isin(netw_order[clusterings[n_thr][2]]))] = int(3)
    
    
    
    for hemi in ['rh', 'lh']:
        print '\n%s' % hemi


        data = np.array(df['network clusters'][(df['threshold'] == thr) & (df['hemisphere'] == hemi)])

        labels = tuple([df['node'][(df['network clusters'] == x)
                                      & (df['hemisphere'] == hemi)
                                      & (df['threshold'] == thr)] 
                       for x in [1,2,3]])

        create_fig(labels=labels, hemi=hemi, surf='inflated', 
                   file_name='/scr/hunte1/sabine/distconnect/figures/images/suppl_netw_clusters_%s_%s' 
                   % (thr, hemi),
                   cpal=sns.xkcd_palette(["bluish purple", "orange", "dandelion"]))
```
